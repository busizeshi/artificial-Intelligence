数组的连接

```python
a=np.array([[1,2,3],[4,5,6]])
b=np.array([[7,8,9],[10,11,12]])
```

```python
c=np.concatenate((a,b))
```

```python
c
```

```python
c=np.concatenate((a,b),axis=1)
c
```

```python
c=np.concatenate((a,b),axis=0)
```

```python
c
```

```python
np.vstack((a,b))
```

```python
np.hstack((a,b))
```
