保存numpy数据

```python
import numpy as np
```

```python
x=np.arange(0,10,dtype=np.int16).reshape((2,5))
x
```

```python
np.savetxt("jwd.txt", x,fmt="%d",delimiter=";")
```

读写array结构

```python
x=np.arange(0,10,dtype=np.int16).reshape((2,5))
np.save("jwd.npy",x)
```

```python
y=np.load("jwd.npy")
y
```

```python
x=np.arange(0,10,dtype=np.int16).reshape((2,5))
y=np.arange(10,20,dtype=np.int16).reshape((2,5))
np.savez("jwd.npz",x=x,y=y)
```

```python
a=np.load("jwd.npz")
a.keys()
```
