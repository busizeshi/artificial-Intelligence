# pandas功能

```python
import pandas as pd
```

## 基本功能

```python
# 重新索引
s=pd.Series([10,20,30],index=['a','b','c'])
s.reindex(['b','c','a'])
```

```python
s.reindex(['b','c','a','d'])
```

```python
s.reindex(['b','c','a','d'],fill_value=0)
```

```python
# 丢弃指定轴上的项
df=pd.DataFrame([[1,2],[3,4]],index=['a','b'],columns=['c','d'])
df
```

```python
# 删除行
df.drop('a')
```

```python
# 删除列
df.drop('c',axis=1)
```

```python
df.drop(columns=['c'])
```

```python
# 原地修改
df.drop('a',inplace=True)
```

```python
df
```

```python
# 算数运算和数据对齐
s1 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s2 = pd.Series([10, 20], index=['b', 'c'])
s1
```

```python
s2
```

```python
s1+s2
```

```python
# 填充缺失值
s1.add(s2,fill_value=0)
```

```python
# 函数应用和映射
df[['Age']].apply(sum)
```

```python
s = pd.Series([1, 2, 3])
s.map(lambda x: x * 10)
```

```python
# 排序和排名
df.sort_index(axis=1)
```

```python
df.sort_values(by='Age', ascending=False)
```
