```python
# 处理缺失数据
import pandas as pd
```

```python
df=pd.read_csv('../../data/titanic.csv')
```

```python
# 每列缺失数量
df.isna().sum()
```

```python
# 缺失比例
df.isna().mean().sort_values(ascending=False)
```

```python
# 删除缺失数据
df.dropna()
```

```python
# 删除某一列
df.drop(columns=["Cabin"])
```

```python
# 只有有空值就删行
df.dropna(how="any")
```

```python
# 用常数填充
df["Age"].fillna(0)
```

```python
# 均值填充
df["Age"].fillna(df["Age"].mean())
```

```python
# 中位数填充
df["Age"].fillna(df["Age"].median())
```

```python
# 用众数填充
df["Embarked"].mode()
```

```python
df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
# 层次化索引
survival = df.groupby(["Sex", "Pclass"])["Survived"].mean()
survival
```
